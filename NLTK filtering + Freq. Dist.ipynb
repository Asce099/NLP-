```python
#Stopword filtering bloc
import nltk
from nltk.corpus import stopwords
```

```python
stops1 = set(stopwords.words('english'))
print(stops1)
len(stops1)
```

```python
words = word_tokenize(gunviob)
wordsFiltered = []
```

```python
for w in words:
    if w not in stops1:
        wordsFiltered.append(w)

print(wordsFiltered)
```
